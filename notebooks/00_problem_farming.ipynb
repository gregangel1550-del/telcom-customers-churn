### Objective
Predict which telecom customers will cancel their subscription whithin the next 30 days so that team can intervene with a target offer.

this is a **binary calssification** problem:
- `Churn = 1` -> customer left within 30 days
- `Churn = 0` -> customer stayed

### AUC-ROC as the primary metric

The target class is imbalance (~26% churn). Using accuracy would be missleading, a model predicting 'no churn' for everyone score 74% accuracy while being useless.

AUC-ROC measures the model's ability to rank a churner above a non-churner across all decision therholds. It is robust to class imbalance and gives a single number that reflects overall discriminative power.

**Target:** AUC-ROC >= 0.82 on the held-out test set. 

### Decision thershold 

The default classification thershold (0.5) is not used. we lower it to **0.4** because:
- A false negative (missed churner) means lost revenue - the customer leaves
- A false positive (non-churner flagged) costs only a small retention offer

At threshold  0.4, we accept more false positive in exchange for catching more real churners. The final threshold will be tuned after model training using the precision-recall curve.

```python
monthly_customers = 7043
churn_rate = 0.26
avg_monthly_revenue = 65 # USD, from dataset MonthlyCharge column

monthly_churners = monthly_customers * churn_rate
monthly_revenue_lost = monthly_churners * avg_monthly_revenue

#If model recovers 40% of churners and retention converts 50% of those
model_recovery_rate = 0.40
conversion_rate = 0.50
monthly_saving = monthly_churners * model_recovery_rate * conversion_rate *avg_monthly_revenue


print(f"Estimated monthly churners:      {monthly_churners:.0f} customers")
print(f"Monthly revenue lost to churn:   ${monthly_revenue_lost:,.0f}")
print(f"Projected monthly saving:        ${monthly_saving:,.0f}")
```
